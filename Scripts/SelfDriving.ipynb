Training the model with YOLOV8 medium

```python
if torch.cuda.is_available():
    device = torch.device('cuda')
    print("GPU is available. Using GPU:", torch.cuda.get_device_name(0))
else:
    device = torch.device('cpu')
    print("GPU not available. Using CPU.")

yaml_file = '/content/drive/MyDrive/AI_Folder/Datasets/SelfDriving_Dataset/data.yaml'
weights = 'yolov8m.pt'

model = YOLO(weights)

model.train(
    data=yaml_file,
    epochs=25,
    patience = 10,
    batch = 16,
    imgsz=800,
    project='/content/drive/MyDrive/AI_Folder/Models/SelfDriving_CV',
    name='self_driving_CV',
)

results = model.val()
print("Validation Results:", results)
```
